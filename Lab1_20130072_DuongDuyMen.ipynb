```python
# Task 1.1

def create_list():
  n = int(input('Enter element quantity: '))
  l = []
  for i in range(n):
    element_value = int(input(f'value of index {i} = '))
    l.append(element_value)
  return l


def min(list):
  min = list[0]
  for i in list:
    if i < min:
      min = i
  return min


def max(list):
  max = list[0]
  for i in list:
    if i > max:
      max = i
  return max


def sum(list):
  sum = 0
  for i in list:
    sum+=i
  return sum


def count_positive_negative(lst):
    positive_count = 0
    negative_count = 0

    for num in lst:
        if num > 0:
            positive_count += 1
        elif num < 0:
            negative_count += 1

    return positive_count, negative_count


list = create_list()
print('list = ', list)

max = max(list)
print('max = ', max)

min = min(list)
print('min = ',min)

sum = sum(list)
print('sum = ',sum)

list.sort();
print('list after sorted = ', list)

posi, nega = count_positive_negative(list)
print('posi numbers = ' ,posi)
print('nega numbers = ' ,nega)




```
